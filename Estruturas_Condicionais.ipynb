#Sistema de verificação de idade para entrada em eventos

```python

def verificar_idade():
    print(" Sistema de Verificação de Idade para Eventos ")

    # Entrada de dados
    idade_input = input("Digite sua idade: ")

    # Verificação se a entrada é numérica
    if not idade_input.isdigit():
        print("Erro: Digite apenas números inteiros para a idade.")
        return  # Encerra a função em caso de erro

    idade = int(idade_input)  # Conversão para inteiro após validação

    # Estrutura condicional para verificar a idade
    if idade < 0:
        print("Idade inválida! A idade não pode ser negativa.")
    elif idade < 12:
        print("Entrada proibida: evento permitido apenas para maiores de 12 anos.")
    elif idade < 18:
        print("Entrada permitida somente com responsável.")
    else:
        print("Entrada liberada! Bem-vindo ao evento.")

# Chamada da função principal
verificar_idade()
```

Aqui estamos criando uma função chamada verificar_idade().
Funções servem para organizar o código.

input() serve para receber informação do usuário.
O que ele digitar será guardado na variável idade_input.

O comando .isdigit() verifica se o texto digitado contém somente números.
Se não for número vai mostra erro
return encerra a função (não deixa o programa continuar)

verificar_idade() - Faz o programa rodar.

input()	Pega a idade do usuário
isdigit()	Confere se é número
int()	Converte para número inteiro
if / elif / else	Decide se pode entrar no evento
print()	Mostra mensagens ao usuário
